## Relative Humidity

```python

import plotly.graph_objects as go

import pandas as pd

df = pd.read_csv("sensors-1-2-10-25-26-27.csv")
df["timestamp"] = pd.to_datetime(df['timestamp'])
 
RHtrace1 = go.Scatter(customdata=df, 
                    y=df['h1'], 
                    x = df['timestamp'], 
                    mode='lines', 
                    hoverinfo='all', 
                    name='1',
                    )
RHtrace2 = go.Scatter(customdata=df, 
                    y=df['h2'], 
                    x = df['timestamp'], 
                    mode='lines', 
                    hoverinfo='all', 
                    name='2',
                    )

RHtrace10 = go.Scatter(customdata=df, 
                    y=df['h10'], 
                    x = df['timestamp'], 
                    mode='lines', 
                    hoverinfo='all', 
                    name='10',
                    )
RHtrace25 = go.Scatter(customdata=df, 
                    y=df['h25'], 
                    x = df['timestamp'], 
                    mode='lines', 
                    hoverinfo='all', 
                    name='25',
                    )

RHtrace26 = go.Scatter(customdata=df, 
                    y=df['h26'], 
                    x = df['timestamp'], 
                    mode='lines', 
                    hoverinfo='all', 
                    name='26',
                    )
RHtrace27 = go.Scatter(customdata=df, 
                    y=df['h27'], 
                    x = df['timestamp'], 
                    mode='lines', 
                    hoverinfo='all', 
                    name='27',
                    )
RHtraceLascar = go.Scatter(customdata=df, 
                    y=df['Lascar-RH'], 
                    x = df['timestamp'], 
                    mode='lines', 
                    hoverinfo='all', 
                    name='Lascar',
                    )

g = go.FigureWidget(data=[RHtrace1, RHtrace2, RHtrace10, RHtrace25, RHtrace26, RHtrace27,  RHtraceLascar])
g.layout.title = 'RH readings - sensors 1, 2, 10, 25-27'
g.layout.xaxis.title= 'timestamp'
g.layout.yaxis.title = "%RH"


fig = go.Figure(g)

fig.show()

```

## Temperature

```python


temptrace1 = go.Scatter(customdata=df, 
                    y=df['t1'], 
                    x = df['timestamp'], 
                    mode='lines', 
                    hoverinfo='all', 
                    name='1',
                    )
temptrace2 = go.Scatter(customdata=df, 
                    y=df['t2'], 
                    x = df['timestamp'], 
                    mode='lines', 
                    hoverinfo='all', 
                    name='2',
                    )

temptrace10 = go.Scatter(customdata=df, 
                    y=df['t10'], 
                    x = df['timestamp'], 
                    mode='lines', 
                    hoverinfo='all', 
                    name='10',
                    )
temptrace25 = go.Scatter(customdata=df, 
                    y=df['t25'], 
                    x = df['timestamp'], 
                    mode='lines', 
                    hoverinfo='all', 
                    name='25',
                    )

temptrace26 = go.Scatter(customdata=df, 
                    y=df['t26'], 
                    x = df['timestamp'], 
                    mode='lines', 
                    hoverinfo='all', 
                    name='26',
                    )
temptrace27 = go.Scatter(customdata=df, 
                    y=df['t27'], 
                    x = df['timestamp'], 
                    mode='lines', 
                    hoverinfo='all', 
                    name='27',
                    )
temptraceLascar = go.Scatter(customdata=df, 
                    y=df['Lascar-temp'], 
                    x = df['timestamp'], 
                    mode='lines', 
                    hoverinfo='all', 
                    name='Lascar',
                    )

g = go.FigureWidget(data=[temptrace1, temptrace2, temptrace10, temptrace25, temptrace26, temptrace27,temptraceLascar])
g.layout.title = 'Temperature readings - sensors 1, 2, 10, 25-27'
g.layout.xaxis.title= 'timestamp'
g.layout.yaxis.title = "deg C"


fig = go.Figure(g)

fig.show()

```
